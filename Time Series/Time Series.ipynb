```python
from statsmodels.tsa.ar_model import AutoReg
from random import random
```

```python
#contrived dataset
data = [x + random() for x in range(1, 100)]
```

```python
#fit model

model = AutoReg(data, lags=1)
model_fit = model.fit()
```

```python
#make prediction
yhat = model_fit.predict(len(data), len(data))
print(yhat)
```
